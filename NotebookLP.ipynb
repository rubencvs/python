```python
def string_match(a, b):
  count = 0
  for i in range(len(a)-1):   
    for j in range(len(b)-1): 
      if a[i:i+2] == b[j:j+2]:
        count += 1
  return count
```

```python
def make_bricks(small, big, goal):
  veces = goal / 5
  resto = goal % 5
  
  if small > 5:
    big += small / 5
    small = small % 5
    if (big > veces):
      small += (big - veces) * 5
      big -= big - veces 
    
  
  return (big >= veces and small >= resto) 
```

```python
make_bricks(7, 1, 8)
```

```python
make_bricks(1, 4, 12)
```

```python
def lone_sum(a, b, c):
  sum = 0
  if a != b and a != c: sum += a
  if b != a and b != c: sum += b
  if c != a and c != b: sum += c
  
  return sum
```

```python
lone_sum(1, 2, 3) 
```

```python
lone_sum(2, 2, 9)
```

```python
def round_sum(a, b, c):
  return round10(a) + round10(b) + round10(c)
  
  
def round10(num):
  resto = num % 10
  diferencia = 10 - resto
  
  if resto >= diferencia:
    return num + diferencia 
  else:
    return num - resto
```

```python
round_sum(16, 17, 18)
```

```python
round_sum(12, 15, 14) 
```

```python
def count_hi(str):
  count = 0
  for i in range(len(str)-1):
    if (str[i]=="h" and str[i+1]=="i"):
      count += 1
  
  return count 
```

```python
def cat_dog(str):
  cat=0
  dog=0
  
  for i in range(len(str)-2):
    if str[i:i+3] == "cat":
      cat += 1
      
    if str[i:i+3] == "dog":
      dog += 1
  
  return cat==dog 
```

```python
def end_other(a, b):
  la = len(a)
  lb = len(b)
  
  ml = max(la,lb)
  
  if la==ml:
    return (a[-lb:].lower() == b.lower())
  else:
    return (b[-la:].lower() == a.lower())
```

```python
end_other('xyz', '12xyz')
```

```python
def count_evens(nums):
  counte = 0
  
  for i in range(len(nums)):
    if (nums[i] % 2 == 0):
      counte += 1
  
  return counte   
```

```python
def big_diff(nums):
  minN = nums[0]
  maxN = nums[0]
  
  for i in range(len(nums)):
    minN = min(minN,nums[i])
    maxN = max(maxN,nums[i])
    
  return maxN-minN
```

```python
big_diff([5, 1, 6, 1, 9, 9]) 
```

```python
def centered_average(nums):
  minN = min(nums[0],nums[len(nums)-1])
  maxN = max(nums[0],nums[len(nums)-1])
  sumaN = 0
 
  for i in range(len(nums)-2):
    
    if (nums[i+1] < minN):
      sumaN += minN
      minN = nums[i+1] 
    elif (nums[i+1] > maxN):
      sumaN += maxN
      maxN = nums[i+1]
    else:
      sumaN += nums[i+1]  
           
  return int(sumaN / (len(nums)-2))    
```

```python
centered_average([100, 0, 5, 3, 4])
```

```python
def sum13(nums):
  suma=0
  after13 = False
  
  for i in range(len(nums)):
    if nums[i] != 13 and not(after13):
      suma += nums[i]
      after13 = False
    elif nums[i] == 13:
      after13 = True
    else:
      after13 = False
      
  return suma  
```

```python
sum13([13, 1, 2, 13, 2, 1, 13]) 
```

```python
def sum67(nums):
  suma=0
  started6 = False
  
  for i in range(len(nums)):
    if nums[i] != 6 and not(started6):
      suma += nums[i]
    elif nums[i] == 6:
      started6 = True
    elif nums[i] == 7:
      started6 = False
      
  return suma   
```

```python
def has22(nums):
  for i in range(len(nums)-1):
    if nums[i] == 2 and nums[i+1] == 2:
      return True
  
  return False  
```
